```python
import chart_studio.plotly as py
import plotly.graph_objs as go
import pandas as pd
from plotly.offline import plot
df1 = pd.read_csv("iris.csv")
df1.head()

data = []
clusters = []
colors = ['rgb(228,26,28)', 'rgb(55,126,184)', 'rgb(77,175,74)']

for i in range(len(df1['variety'].unique())):
    name = df1['variety'].unique()[i]
    color = colors[i]
    x = df1[df1['variety']==name]['sepal.length']
    y = df1[df1['variety']==name]['sepal.width']
    z = df1[df1['variety']==name]['petal.length']

    trace = dict(
        name = name,
        x = x, y=y ,z = z,
        type = "scatter3d",
        mode = 'markers',
        marker = dict(size = 3, color = color , line = dict(width = 0 )))
    data.append(trace)

layout = dict(
    width  = 800,
    height = 550,
    autosize = False,
    title = "Iris dataset",
    scene = dict(
        xaxis= dict(
            gridcolor = 'rgb(255,255,255)',
            zerolinecolor = 'rgb(255,255,255)',
            showbackground = True,
            backgroundcolor = 'rgb(230,230,230)'
        ),
        yaxis= dict(
            gridcolor = 'rgb(255,255,255)',
            zerolinecolor = 'rgb(255,255,255)',
            showbackground = True,
            backgroundcolor = 'rgb(230,230,230)'
        ),
        zaxis= dict(
            gridcolor = 'rgb(255,255,255)',
            zerolinecolor = 'rgb(255,255,255)',
            showbackground = True,
            backgroundcolor = 'rgb(230,230,230)'
         ),
         aspectratio = dict(x=1 , y = 1, z= 0.7),
         aspectmode = 'manual'
    ),
  
)

fig = dict(data= data, layout = layout)

plot(fig, filename='iris.html', auto_open=False)
```

```python
import numpy as np 
counts , bin_edges = np.histogram(iris['petal.length'], bins = 10, density = True )
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)
print(counts)

cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:] , cdf)

plt.show()
```

## Mean , Variance and Std-dev

```python
print("Means:")
print(np.mean(iris["petal.length"]))

print(np.mean(np.append(iris["petal.length"],50)))
```

```python
print("\nStd-dev:")
print(np.std(iris['petal.length']))
```

## Median , percentile , Quantile , IQR , MAD

```python
pip install statsmodels
```

```python
print("\nMedians")
print(np.median(iris['petal.length']))
print(np.median(np.append(iris['petal.length'] , 50)))


print("\nQuantiles:")
print(np.percentile(iris['petal.length'], np.arange(0,100,25)))

print("\n90th Percentiles:")
print(np.percentile(iris['petal.length'],90))

from statsmodels import robust
print ("\n Median absolute  Deviation")
print(robust.mad(iris["petal.length"]))
```
